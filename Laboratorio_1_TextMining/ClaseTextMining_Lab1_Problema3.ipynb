```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# leemos nuestra imagen
img = cv2.imread("Camila.JPG")
```

```python
# graficamos para validar la imagen
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
# obtenemos las dimensiones de nuestra imagen
alto = img.shape[0]
ancho = img.shape[1]
```

```python
# creamos una matriz por cada imagen
lienzo_azul = np.zeros((alto, ancho, 1))
lienzo_verde = np.zeros((alto, ancho, 1))
lienzo_rojo = np.zeros((alto, ancho, 1))
```

```python
# creamos nuestra funcion que nos retorna 3 graficas, azul, verde y rojo.
def img_grises (img):
    for i in range (0, alto):
        for j in range(0, ancho):
            pixel = img[i,j]
            
            blue = pixel[0]
            green = pixel[1]
            red = pixel[2]
            lienzo_azul[i,j] = blue 
            lienzo_verde[i,j] = green 
            lienzo_rojo[i,j] = red
            
    cv2.imwrite ("monocromo_azul.jpg", lienzo_azul)
    monocromo_azul = cv2.imread("monocromo_azul.jpg")
    monocromo_azul = cv2.cvtColor(monocromo_azul, cv2.COLOR_BGR2RGB)

    cv2.imwrite ("monocromo_verde.jpg", lienzo_verde)
    monocromo_verde = cv2.imread("monocromo_verde.jpg")
    monocromo_verde = cv2.cvtColor(monocromo_verde, cv2.COLOR_BGR2RGB)

    cv2.imwrite ("monocromo_rojo.jpg", lienzo_rojo)
    monocromo_rojo = cv2.imread("monocromo_rojo.jpg")
    monocromo_rojo = cv2.cvtColor(monocromo_rojo, cv2.COLOR_BGR2RGB)
    
    return monocromo_azul,monocromo_verde, monocromo_rojo
        
```

```python
# Obtenemos y graficamos la imagen azul de la funcion
img_azul = img_grises(img)[0]
plt.imshow(img_azul)
plt.show()
```

```python
# Obtenemos y graficamos la imagen verde de la funcion
img_verde = img_grises(img)[1]
plt.imshow(img_verde)
plt.show()
```

```python
# Obtenemos y graficamos la imagen roja de la funcion
img_rojo = img_grises(img)[2]
plt.imshow(img_rojo)
plt.show()
```
