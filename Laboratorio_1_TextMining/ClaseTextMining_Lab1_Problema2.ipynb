```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# leemos nuestras imagenes
img1_azul = cv2.imread("perro_salida_gray_azul.jpg")
img2_verde = cv2.imread("perro_salida_gray_verde.jpg")
img3_rojo = cv2.imread("perro_salida_gray_rojo.jpg")
```

```python
# conversion de BGR a RGB
img1_azul = cv2.cvtColor(img1_azul, cv2.COLOR_BGR2RGB)
img2_verde = cv2.cvtColor(img2_verde, cv2.COLOR_BGR2RGB)
img3_rojo = cv2.cvtColor(img3_rojo, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1_azul)
plt.show()
```

```python
plt.imshow(img2_verde)
plt.show()
```

```python
plt.imshow(img3_rojo)
plt.show()
```

```python
# obtenemos la dimension de nuestra imagen, puede ser cualquiera de las 3, siempre y cuando sean 
# las misma grafia y dimensiones en diferente tono de grises
img1_azul.shape
```

```python
# estas dimensiones son generales
alto = img1_azul.shape[0]
ancho = img1_azul.shape[1]
```

```python
ancho
```

```python
alto
```

```python
# generamos nuestra matris donde vamos a guardar los resultados de las 3 graficas para mostrar la grafica original
lienzo = np.zeros((alto, ancho, 3))
```

```python
for i in range (0, alto):
    for j in range (0,ancho):
        # aca recorremos todos los pixeles azules de imagen
        pixel_azul = img1_azul[i,j]
        pixel_verde = img2_verde[i,j]
        pixel_rojo = img3_rojo[i,j]
        
        blue = pixel_azul[0]
        green = pixel_verde[1]
        red = pixel_rojo[2]
                         
        lienzo[i,j] = [blue, green, red]               
```

```python
cv2.imwrite ("perro_original.jpg", lienzo)
perro_original = cv2.imread("perro_original.jpg")
perro_original = cv2.cvtColor(perro_original, cv2.COLOR_BGR2RGB)
```

```python
# Graficamos la imagen Original
plt.imshow(perro_original)
plt.show()
```
