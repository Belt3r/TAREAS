```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# leemos nuestra imagen
img = cv2.imread("Camila.JPG")
```

```python
# convertimos a RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
alto = img.shape[0]
ancho = img.shape[1]
```

```python
# sacamos nuestra matris
media_ponderada = np.zeros((alto, ancho, 1))
```

```python
# definimos nuestra funcion
color = ('b','g','r')
def img_histo(img):
    for i, col in enumerate(color):
        histr = cv2.calcHist([img],[i],None,[256],[0,256])
        plt.plot(histr,color = col)
    
    for i in range(0, alto):
        for j in range (0, ancho):
            pixel = img [i,j]
            blue = pixel[0]
            green = pixel[1]
            red = pixel[2]
        
            #media_ponderada[i,j] = int ((0.114*blue + 0.587*green + 0.299*red)/3)
            media_ponderada[i,j] = np.average(pixel)
    cv2.imwrite("gris_ponderado.jpg", media_ponderada)
    gris = cv2.imread("gris_ponderado.jpg")
    gris = cv2.cvtColor(gris, cv2.COLOR_BGR2RGB)
    histr_2 = cv2.calcHist([gris], [0], None, [256], [0, 256])
    plt.plot(histr_2, color='gray' )
    return histr, histr_2
```

```python
# graficamos los 4 histogramas
img_histo(img)
plt.xlim([0,256])
plt.show()
```
