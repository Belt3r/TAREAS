```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## PROBLEMA # 1

```python
# ingreso de parametros
img = cv2.imread("Camila.JPG")
color = 3
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
# obtenemos las dimensiones de nuestra imagen
alto = img.shape[0]
ancho = img.shape[1]
```

```python
# creamos nuestra matris, donde vamos a guardar el resultado, esta debe de ser de las mismas dimensiones
# que la original.
lienzo = np.zeros((alto, ancho, 3))
```

```python
# funcion para parametros de imagen y color
def imagen(img, color):
    if color == 1:
        for i in range (0, alto):
            for j in range(0, ancho):
                pixel = img[i,j]
                blue = pixel[0]
                green = pixel[1]
                red = pixel[2]
                lienzo[i,j] = [blue,0,0]
        cv2.imwrite ("monocromo.jpg", lienzo)
        monocromo = cv2.imread("monocromo.jpg")
        monocromo = cv2.cvtColor(monocromo, cv2.COLOR_BGR2RGB)
    if color == 2:
        for i in range (0, alto):
            for j in range(0, ancho):
                pixel = img[i,j]
                blue = pixel[0]
                green = pixel[1]
                red = pixel[2]
                lienzo[i,j] = [0,green,0]
        cv2.imwrite ("monocromo.jpg", lienzo)
        monocromo = cv2.imread("monocromo.jpg")
        monocromo = cv2.cvtColor(monocromo, cv2.COLOR_BGR2RGB)
    if color == 3:
        for i in range (0, alto):
            for j in range(0, ancho):
                pixel = img[i,j]
                blue = pixel[0]
                green = pixel[1]
                red = pixel[2]
                lienzo[i,j] = [0,0,red]
        cv2.imwrite ("monocromo.jpg", lienzo)
        monocromo = cv2.imread("monocromo.jpg")
        monocromo = cv2.cvtColor(monocromo, cv2.COLOR_BGR2RGB)
    if color == 10:
        for i in range (0, alto):
            for j in range(0, ancho):
                pixel = img[i,j]
                blue = pixel[0]
                green = pixel[1]
                red = pixel[2]
                lienzo[i,j] = [0,green,red]
        cv2.imwrite ("monocromo.jpg", lienzo)
        monocromo = cv2.imread("monocromo.jpg")
        monocromo = cv2.cvtColor(monocromo, cv2.COLOR_BGR2RGB)
    if color == 20:
        for i in range (0, alto):
            for j in range(0, ancho):
                pixel = img[i,j]
                blue = pixel[0]
                green = pixel[1]
                red = pixel[2]
                lienzo[i,j] = [blue,green,0]
        cv2.imwrite ("monocromo.jpg", lienzo)
        monocromo = cv2.imread("monocromo.jpg")
        monocromo = cv2.cvtColor(monocromo, cv2.COLOR_BGR2RGB)
    if color == 30:
        for i in range (0, alto):
            for j in range(0, ancho):
                pixel = img[i,j]
                blue = pixel[0]
                green = pixel[1]
                red = pixel[2]
                lienzo[i,j] = [blue,0,red]
        cv2.imwrite ("monocromo.jpg", lienzo)
        monocromo = cv2.imread("monocromo.jpg")
        monocromo = cv2.cvtColor(monocromo, cv2.COLOR_BGR2RGB)
    return monocromo
              
```

```python
# mostramos la imagen
plt.imshow(imagen(img, color))
plt.show()
```
